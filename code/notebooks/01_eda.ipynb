```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sys
from pathlib import Path

# Add the 'code' directory to the system path so we can import our config
sys.path.append(str(Path.cwd().parent))
from config import DATA_DIR

def plot_s_parameters(filename):
    file_path = DATA_DIR / filename
    if not file_path.exists():
        print(f"Error: Could not find {file_path}")
        return

    # Load the dataset
    df = pd.read_csv(file_path)
    print(f"Dataset Loaded: {filename}")
    print(f"Shape: {df.shape}")
    display(df.head())

    # Basic Visualization of S11 and S21 if they exist
    # Adjust column names ('Frequency', 'S11_mag', etc.) based on your actual CSV headers
    freq_col = [col for col in df.columns if 'freq' in col.lower()]

    if freq_col:
        x_axis = df[freq_col[0]]
        plt.figure(figsize=(12, 6))

        # Plot S11
        s11_cols = [col for col in df.columns if 's11' in col.lower()]
        for col in s11_cols:
            plt.plot(x_axis, df[col], label=col, alpha=0.8)

        # Plot S21
        s21_cols = [col for col in df.columns if 's21' in col.lower()]
        for col in s21_cols:
            plt.plot(x_axis, df[col], label=col, linestyle='--', alpha=0.8)

        plt.title('Microwave S-Parameters vs. Frequency')
        plt.xlabel('Frequency')
        plt.ylabel('Magnitude (dB)')
        plt.legend()
        plt.grid(True)
        plt.show()
    else:
        print("Could not automatically detect a Frequency column. Please verify your column names.")

# Replace 'your_data_file.csv' with an actual file in your Datafiles folder
# plot_s_parameters('your_data_file.csv')
```
